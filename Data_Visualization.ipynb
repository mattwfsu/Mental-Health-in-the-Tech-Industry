```python
#drop rows with NA values
dfCombinedFinal = dfCombinedFinal.dropna(how='any')
dfCombinedFinal
```

```python
#Create a histogram of each variable
for col in dfCombinedFinal.columns:
    plt.figure(figsize=(4, 3))  # create a larger figure
    plt.hist(dfCombinedFinal[col], bins=10, color='skyblue', alpha=0.75)  # customize color and transparency
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.75)  # add a grid
    plt.xticks(rotation=45)
    plt.show()
```

```python
#Create a heatmap of the correlation matrix of the dataframe
plt.figure(figsize = (38,16))
sns.heatmap(dfCombinedVectorizedna.corr(), annot = True)
plt.savefig('heatmap.png')
plt.show()
```
